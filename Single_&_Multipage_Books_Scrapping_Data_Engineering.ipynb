```python
from bs4 import BeautifulSoup
import requests
from csv import writer
import pandas as PD

url= "https://www.amazon.in/s?k=books+on+geospatial+data&ref=nb_sb_noss"
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}
page = requests.get(url, headers=headers)

soup_ = BeautifulSoup(page.content, 'html.parser')
soup = BeautifulSoup(soup_.prettify(), "html.parser")

containers = soup.findAll("div",{"class":"a-section a-spacing-small a-spacing-top-small"})
print(len(containers))

with open('Geobooks.csv', 'w', encoding='utf8', newline='') as f:
    thewriter = writer(f)
    header = ['Title','Book_Types','Price']
    thewriter.writerow(header)
    
    for container in containers:
        titles = container.find_all("span",class_="a-size-medium a-color-base a-text-normal")
        for title in titles:
            title_header = title.text.strip()
            print(title_header)
            booktypes = container.find_all("a",class_="a-size-base a-link-normal s-underline-text s-underline-link-text s-link-style a-text-bold")
            for booktype in booktypes:
                booktype_header = booktype.text.strip()
                prices = container.find_all("span",class_="a-price-whole")
                for price in prices:   
                    price_header = price.text.strip()
                    print(price_header)
                    info = [title_header,booktype_header,price_header]
                    thewriter.writerow(info)
```

```python
from bs4 import BeautifulSoup
import requests
from csv import writer
import pandas as PD

Geobooks_multipages = []

for i in range(1,5):
    
    url= f"https://www.amazon.in/s?k=books+on+geospatial+data&page={i}&qid=1675200402&ref=sr_pg_{i}"
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}
    page = requests.get(url, headers=headers)

    soup_ = BeautifulSoup(page.content, 'html.parser')
    soup = BeautifulSoup(soup_.prettify(), "html.parser")

    containers = soup.findAll("div",{"class":"a-section a-spacing-small a-spacing-top-small"})
    print(len(containers))

    for container in containers:
        titles = container.find_all("span",class_="a-size-medium a-color-base a-text-normal")
        for title in titles:
            title_header = title.text.strip()
            booktypes = container.find_all("a",class_="a-size-base a-link-normal s-underline-text s-underline-link-text s-link-style a-text-bold")
            for booktype in booktypes:
                booktype_header = booktype.text.strip()
                prices = container.find_all("span",class_="a-price-whole")
                for price in prices:   
                    price_header = price.text.strip()
                    Geobooks_multipages.append([title_header,booktype_header,price_header])
                    #print(Geobooks_multipages)
                    
                    
                    
df = PD.DataFrame(Geobooks_multipages, columns=['Title','Book_Types','Price'])
print(df)
df.to_csv('Geobooks_multipages.csv')                   
```
